```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
def vader_sentiment_analysis(text):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_score = analyzer.polarity_scores(text)
    return sentiment_score['compound']
```

```python
def textblob_sentiment_analysis(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity
```

```python
def main():
    print("Loading preprocessed data...")

    df = pd.read_csv('preprocess_reviews.csv')

    print("Performing sentiment analysis using VADER...")
    df['vader_sentiment'] = df['reviewText'].apply(lambda x: vader_sentiment_analysis(str(x)))
    df['textblob_sentiment'] = df['reviewText'].apply(lambda x: textblob_sentiment_analysis(str(x)))  

    def categorize_vader_sentiment(score):
        if score >= 0.05:
            return 'positive'
        elif score <= -0.05:
            return 'negative'
        else:
            return 'neutral'
    
    df['vader_sentiment_category'] = df['vader_sentiment'].apply(categorize_vader_sentiment)
    df.to_csv('sentiment_analysis_results.csv', index=False)
    print("Sentiment analysis completed. Results saved to 'sentiment_analysis_results.csv'.")
    print("Sentiment distribution:")
    print(df['vader_sentiment_category'].value_counts())

if __name__ == "__main__":
    main()
```
