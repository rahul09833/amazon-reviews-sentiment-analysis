```python

import pandas as pd
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import numpy as np
import os
```

```python
def main():
    print("Loading Word2Vec model...")
    model = Word2Vec.load("models/word2vec.model")
    
    # Select words to visualize (most frequent or interesting ones)
    # We'll take the top 100 most frequent words from the model's vocabulary
    words = list(model.wv.index_to_key)[:50]
    word_vectors = np.array([model.wv[word] for word in words])
    
    print("Reducing dimensions with t-SNE...")
    # Reduce 100D vectors to 2D for plotting
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(30, len(words)-1))
    vectors_2d = tsne.fit_transform(word_vectors)
    
    # Plotting
    plt.figure(figsize=(15, 10))
    plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.6, edgecolors='k')
    
    for i, word in enumerate(words):
        plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]), fontsize=12, xytext=(5, 2), textcoords='offset points')

        
    plt.title('t-SNE Visualization of Word2Vec Word Embeddings(Top 50 Words)', fontsize=16)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Create plots directory if it doesn't exist
    os.makedirs('plots', exist_ok=True)
    plt.savefig('plots/word2vec_tsne.png')
    plt.close()
    print("t-SNE plot saved to 'plots/word2vec_tsne.png'.")

if __name__ == "__main__":
    main()
```
